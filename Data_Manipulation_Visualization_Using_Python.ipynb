Use numpy.where() to encode gender column.

```python
users['gender_encoded'] = np.where(users['gender'] == 'M', 1, 0)
male_ratio_per_occupation = users.groupby('occupation')['gender_encoded'].mean()
sorted_male_ratio = male_ratio_per_occupation.sort_values(ascending=False)
print("Male ratio per occupation:", sorted_male_ratio)
```

### 33.  For each occupation present the percentage of women and men

```python
gender_count = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)

total_count = gender_count.sum(axis=1)

gender_percentage = gender_count.div(total_count, axis=0) * 100
print("Percentage of women and men for each occupation:",gender_percentage)
```

### 51. Write a Python program to subtract five days from current date.

Current Date : 2015-06-22

5 days before Current Date : 2015-06-17

```python
from datetime import timedelta

current_date = datetime.strptime("2015-06-22", "%Y-%m-%d")

new_date = current_date - timedelta(days=5)
formatted_new_date = new_date.strftime("%Y-%m-%d")

print("Current Date :", current_date.strftime("%Y-%m-%d"))
print("5 days before Current Date :", formatted_new_date)
```

### 80. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries
Hint: First we need to find top-3 countries based on revenue, then create scater plot between Quantity and Unitprice for each country separately


```python
online_rt_filt = online_rt[online_rt['Quantity'] > 0]

online_rt_filt['Revenue'] = online_rt_filt['Quantity'] *online_rt_filt['UnitPrice']

country_revenue = online_rt_filt.groupby('Country')['Revenue'].sum()
top_countries = country_revenue.sort_values(ascending=False).head(3).index

online_rt_top_countries = online_rt_filt[online_rt_filt['Country'].isin(top_countries)]

for country in top_countries:
    df_country = online_rt_top_countries[online_rt_top_countries['Country'] == country]
    plt.figure(figsize=(10, 6))
    plt.scatter(df_country['UnitPrice'], df_country['Quantity'], alpha=0.5, label=country)
    plt.title(f'Quantity vs. UnitPrice for {country}')
    plt.xlabel('UnitPrice')
    plt.ylabel('Quantity')
    plt.legend()
    plt.show()
```
